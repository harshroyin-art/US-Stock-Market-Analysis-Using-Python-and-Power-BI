## Fiancial Return and volatility analysis:

```python
#1) Daily percentage return:
```

#### Method 1:

```python

# Method1:
# Here we will compute the change in percentage of closing price of each day with their previous day for different tickers.
# formula----> [(closing price of current day-closing price of previous day)/closing price of previous day]*100

print("the different tickers are",df["Ticker"].unique())
tick=input("enter the name of the ticker from the above")
# 1) We will make a data frame from the original dataframe for a particular ticker as 'tick' and arrange the data frame 
# in chronological order as per date.
df_2=df[df["Ticker"]==tick]
df_2=df_2.sort_values(["Date"])
# 2) Now we will find out the current day close price
df_2["Current_day_close"]=df["Close"]
# 3) Now we will find out the previous day close price
df_2["Previous_day_close"]=df["Close"].shift(1)
df_2

# Calculation:-
df_2["Daily_return_percentage"]=((df_2["Current_day_close"]-df_2["Previous_day_close"])/df_2["Previous_day_close"])*100
df_2
```

#### Method 2: Using "pct_change" as a function

```python
# 1) We will make a data frame from the original dataframe for a particular ticker as 'tick' and arrange the data frame 
# in chronological order as per date.
print("the different tickers are",df["Ticker"].unique())
tick=input("enter the name of the ticker from the above")
df_3=df[df["Ticker"]==tick]
df_3=df_3.sort_values(["Date"])
df_3['Daily_Return_percentage'] = df_3['Close'].pct_change()
df_3
```

## Advanced analysis:-

```python
# Intraday Price gap and range Analysis( Voltality and Sentiment):-
# Daily closing prices hide what happens during the trading day. we can use the High,Low, and Open columns to calculate
# two vital metrices:-
# 1) Intraday range(High to low): Shows how violently a stock swung back and forth between the opening and closing bells.
# 2) Overnight Gap( Today's Open vs. Yesterday's Close ): Shows how much news accumulated overnight when the market was 
#    closed.
```

```python
# Sort properly across the dataset
df = df.sort_values(['Ticker', 'Date'])

# 1. Intraday Range as a percentage of the Opening price
df['Intraday_Range_Pct'] = (df['High'] - df['Low']) / df['Open']

# 2. Overnight Gap percentage
df['Yesterday_Close'] = df.groupby('Ticker')['Close'].shift(1)
df['Overnight_Gap_Pct'] = (df['Open'] - df['Yesterday_Close']) / df['Yesterday_Close']

# 3. Let's find out which sectors suffer from the widest price gaps or intraday swings
sector_analysis = df.groupby('Sector')[['Intraday_Range_Pct', 'Overnight_Gap_Pct']].mean() * 100
print(sector_analysis)
```

```python
# Sort the original dataset and then store it to a variable as "df_4".
df_4 = df.sort_values(['Ticker', 'Date'])
# Calculation:-
df_4['Intraday_Range_Pct'] = ((df['High'] - df['Low']) / df['Open'])*100
```

```python
# 2. Overnight Gap percentage
df_4['Yesterday_Close'] = df.groupby('Ticker')['Close'].shift(1)
df_4['Overnight_Gap_Pct'] = ((df['Open'] - df_4['Yesterday_Close']) / df_4['Yesterday_Close'])*100
```

```python
df_4
```

```python
# 3. Let's find out which sectors suffer from the widest price gaps or intraday swings
sector_analysis = df_4.groupby('Sector')[['Intraday_Range_Pct', 'Overnight_Gap_Pct']].mean().reset_index()
print(sector_analysis)
```

```python
plt.figure(figsize=(15,5))

df_5=sector_analysis.sort_values(by="Intraday_Range_Pct",ascending=False).reset_index()
plt.subplot(1,2,1)
sns.barplot(x="Sector",y="Intraday_Range_Pct",data=df_5,ci=None,palette="viridis")
plt.title("High to Low fluctuation")
plt.xlabel("Sector")
plt.ylabel("Intraday_Range_Pct")
plt.xticks(rotation=90)

df_5=sector_analysis.sort_values(by="Overnight_Gap_Pct",ascending=False).reset_index()
plt.subplot(1,2,2)
sns.barplot(x="Sector",y="Overnight_Gap_Pct",data=df_5,ci=None,palette="viridis")
plt.title("Overnight Gap")
plt.xlabel("Sector")
plt.ylabel("Overnight_Gap_Pct")
plt.xticks(rotation=90)
plt.show()
```
